### Dataset Information

This is a classification task. A classification problem where we have to predict whether a student would succedd or fail. Below is the dataset attributes with description.

Table 1: The preprocessed student related variables

Attribute	Description (Domain)
 
    school	          student’s school (binary: Gabriel Pereira or Mousinho da Silveira) normalised as 0 or 1
    
    gender	          student’s gender (binary: female or male) normalised as 0 or 1

    age	              student’s age (numeric: from 15 to 22)

    address	          student’s home address type (binary: urban or rural) normalised as 0 or 1

    Pstatus	          parent’s cohabitation status (binary:living together or apart) normalised as 0 or 1

    Medu	          mother’s education (numeric: from 0 to 4)see a
   
    Fedu	          father’s education (numeric: from 0 to 4)see a

    guardian	      student’s guardian (nominal: mother, father or other) normalised as 0, 1 or 2

    famsize           family size (binary: ≤ 3 or > 3) normalised as 0 or 1

    famrel	          quality of family relationships (numeric: from 1 – very bad to 5 – excellent)

    studytime	      weekly study time (numeric: 1 – < 2 hours, 2 – 2 to 5 hours, 3 – 5 to 10 hours or 4 – > 10 hours)

    schoolsup	      extra educational school support (binary: yes or no)

    famsup	          family educational support (binary: yes or no) 

    activities	      extra-curricular activities (binary: yes or no) 

    paidclass	      extra paid classes (binary: yes or no) 

    internet	      Internet access at home (binary: yes or no) 

    nursery	          attended nursery school (binary: yes or no)

    higher	          wants to take higher education (binary: yes or no)

    romantic	      with a romantic relationship (binary: yes or no)

    freetime	      free time after school (numeric: from 1 – very low to 5 – very high)

    goout	          going out with friends (numeric: from 1 – very low to 5 – very high)
    
    abesnces          Number of classes missed per year (numeric)

    G1	              first period grade (numeric: from 0 to 20) 

    G2	              second period grade (numeric: from 0 to 20) 

    G3	              final grade (numeric: from 0 to 20)
    Pass_G3           This is a Binary Classifier with 0 being a fail and 1 being a Pass. see b
 
a	0 – none, 1 – primary education (4th grade), 2 – 5th to 9th grade, 3 – secondary education or 4 – higher education.
b   The mean result of G3 was 11 so all students who received between 0-10 where counted as fails and all who scored 11 or       higher as passes. 
This project aims to use the above data to predict the G3 final grade.

### Feature Engineering

```python
# I am dropping G3 from the dataframe so that it does not cause a fitting issue
df = df.drop(["G3"], axis =1)
```

```python
#Specifiy input and output attribute
X = df.drop(["pass_G3"], axis=1)
Y = df["pass_G3"]
```

```python
#select k best features
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
#apply k-best algorithm
kbest = SelectKBest(score_func = chi2, k = 'all')
ordered_features = kbest.fit(X,Y)
```

```python
df_scores = pd.DataFrame(ordered_features.scores_, columns=["Score"])
```

```python
df_columns = pd.DataFrame(X.columns, columns = ["Feature_name"])
```

```python
feature_rank = pd.concat([df_scores,df_columns], axis=1)
```

```python
feature_rank.nlargest(24, 'Score')
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2, random_state=42)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
def classify(model, x, y):
    x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2, random_state=42)
    model.fit(x_train, y_train)
    print("Accuracy is: ", model.score(x_test, y_test)*100)
    score = cross_val_score(model, x, y, cv=5)
    print("Cross validation Accuracy is: ", np.mean(score)*100)
    
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(solver='liblinear', max_iter=400)
classify(clf, X, Y)
```

```python
from sklearn.svm import SVC
clf = SVC(kernel = 'linear', C=1)
classify(clf, X, Y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5)
classify(knn, X,Y)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
classify(model, X, Y)
modell = model.fit(X,Y)
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
classify(clf, X, Y)
```

```python
clff= clf.fit(X,Y)
```

```python
tree.plot_tree(clff,filled=True)
plt.show()
```

### Confusion Matrix

```python
from sklearn.svm import SVC
clfs = SVC(kernel = 'linear', C=1)
```

```python
from sklearn.linear_model import LogisticRegression
logclf = LogisticRegression(solver='liblinear', max_iter=400)
```

```python
from sklearn.tree import DecisionTreeClassifier

tclf = DecisionTreeClassifier()
```

```python
clfs.fit(x_train, y_train)
```

```python
tclf.fit(x_train, y_train)
```

```python
logclf.fit(x_train, y_train)
```

```python
y_pred = clfs.predict(x_test)
ty_pred = tclf.predict(x_test)
log_pred = logclf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
```

```python
tcm = confusion_matrix(y_test, ty_pred)
lcm = confusion_matrix(y_test, log_pred)
```

```python
#This is a confusion matrix for the SVC model
snsgraph = sns.heatmap(cm, annot=True)
ax = snsgraph
#ax.figure.savefig('confusion_SVC.png') #this line created a local save of the subplot
```

```python
#This is a confusion matrix for the Tree model
snsgraph = sns.heatmap(tcm, annot=True)
ax = snsgraph
#ax.figure.savefig('confusion_tree.png') #this line created a local save of the subplot
```

```python
#This is a confusion matrix for the Logistical regression model
snsgraph = sns.heatmap(lcm, annot=True)
ax = snsgraph
#ax.figure.savefig('confusion_LOG.png') #this line created a local save of the subplot
```

```python
from sklearn.metrics import plot_roc_curve
axc = plot_roc_curve(clfs, x_test, y_test) # SVC
axt = plot_roc_curve(tclf, x_test, y_test) # decision tree
axl = plot_roc_curve(logclf, x_test, y_test) # Logistical regression
```
